```python
import re

def parse_document(document):
    with open(document, 'r', encoding='utf-8') as file:
        data = file.read().replace('\n', ' ')
    
    # Define a pattern for an Annex
    annex_pattern = re.compile(r"Annex [0-9]{1,2}")
    
    # Define a pattern for an Appendix
    appendix_pattern = re.compile(r"Annex [0-9]{1,2} - Appendix [0-9]{1,2}")
    
    # Use findall() to get all matches in the string
    annex_matches = annex_pattern.findall(data)
    appendix_matches = appendix_pattern.findall(data)
    
    # Split the data on Annexes
    annex_splits = re.split(annex_pattern, data)
    
    # Create a dictionary to hold your parsed data
    parsed_data = {}
    
    # Add Annexes to your dictionary
    for i in range(1, len(annex_splits)):
        parsed_data[annex_matches[i-1]] = annex_splits[i]
    
    # Now find and replace any Appendices in the Annexes
    for appendix in appendix_matches:
        for annex in parsed_data.keys():
            if appendix in parsed_data[annex]:
                # Split the Annex on the Appendix
                split_annex = re.split(appendix, parsed_data[annex])
                # Replace the Annex in the dictionary with the part before the Appendix
                parsed_data[annex] = split_annex[0]
                # Add the Appendix to the dictionary
                parsed_data[appendix] = split_annex[1]
    
    return parsed_data
```

```python
document = './r013r6e_Annex.txt'
parsed_data = parse_document(document)
```

```python
for key in parsed_data:
    print(f"{key}: {parsed_data[key]}\n")
    print("\n---------------------------------------------------------------\n")
```
