```python
import pandas as pd

data = {
    'char': [
        'Annex 1\n', 'Data for Annex 1\n', 'Data for Annex 1\n', 'Data for Annex 1\n',
        'Annex 2\n', 'Data for Annex 2\n', 'Data for Annex 2\n', 'Data for Annex 2\n',
        'Annex 2 - Appendix 1\n', 'Data for Annex 2 - Appendix 1\n', 'Data for Annex 2 - Appendix 1\n'
    ]
}

df = pd.DataFrame(data)
```

```python
df['char'] = df['char'].str.rstrip('\n')
```

```python
annex_list = ['Annex 1', 'Annex 2', 'Annex 2 - Appendix 1']

def get_annexes(df, annex_list):
    annexes_content = {}
    current_annex = None
    current_content = []
    
    for index, row in df.iterrows():
        if row['char'] in annex_list:
            if current_annex is not None:
                annexes_content[current_annex] = "\n".join(current_content)
            current_annex = row['char']
            current_content = []
        else:
            if current_annex is not None:
                current_content.append(row['char'])
                
    if current_annex is not None:
        annexes_content[current_annex] = "\n".join(current_content)
    
    return annexes_content
```

```python
annexes_content = get_annexes(df, annex_list)

print(annexes_content)
```

```python
annexes_dict = extract_annexes(df)
```
